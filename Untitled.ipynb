```python
def bldcount(name: str):
    with open(name, 'r') as f:
        line = f.readline()
        
    bloodGroups = line.split()
    
    count_A = 0
    count_B = 0
    count_AB = 0
    count_O = 0
    count_OO = 0

    for bloodGroup in bloodGroups:
        if bloodGroup == 'A':
            count_A+=1
        elif bloodGroup == 'B':
            count_B+=1
        elif bloodGroup == 'AB':
            count_AB+=1
        elif bloodGroup == 'O':
            count_O+=1
        elif bloodGroup == 'OO':
            count_OO+=1 
            
        # exec(bloodGroup + '+=1', globals())

    print(f'Patients of blood type A : {count_A}')
    print(f'Patients of blood type B : {count_B}')
    print(f'Patients of blood type AB : {count_AB}')
    print(f'Patients of blood type O : {count_O}')
    print(f'Patients of blood type OO : {count_OO}')

name = 'bloodtype1.txt'
bldcount(name)
```

Problem 8

```python
def curconv(currency: str, amount: float) -> float:
    with open('currencies.txt', 'r') as f:
        lines = f.readlines()

    for line in lines:
        if line.split()[0] == currency:
            conversionRate = float(line.split()[1])
            amountInUSD = amount * conversionRate
            break
        else:
            continue

    return amountInUSD

print(curconv('AUD', 2345))
```

```python
def frequencies(string: str) -> list:
    string = string.strip()
    string = string.lower()
    alphabets = {'a': 0, 'b': 0, 'c': 0, 'd': 0, 'e': 0, 'f': 0, 'g': 0, 'h': 0, 'i': 0, 'j': 0, 'k': 0, 'l': 0, 'm': 0, 'n': 0, 'o': 0, 'p': 0, 'q': 0, 'r': 0, 's': 0, 't': 0, 'u': 0, 'v': 0, 'w': 0, 'x': 0, 'y': 0, 'z': 0}

    for letter in string:
        if letter in alphabets.keys():
            alphabets[letter] += 1
        else:
            pass
    
    return alphabets.values()

print(frequencies('Apple!!'))
```
